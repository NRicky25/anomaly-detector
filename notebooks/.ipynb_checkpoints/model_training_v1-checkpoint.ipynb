```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
# Set display options for better viewing of DataFrames
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)

csv_file_path = 'data/raw/creditcard.csv'
try:
    df = pd.read_csv(csv_file_path)
    print("Dataset loaded successfully!")
    print(f"Dataframe shape: {df.shape}")
except FileNotFoundError:
    print(f"Error: Dataset not found at {csv_file_path}")
    print("Please ensure 'creditcard.csv' is in the 'data/raw/' directory and the filename is correct.")
    print("Current working directory of the notebook:", os.getcwd()) # This shows where the notebook is running from
except Exception as e:
    print(f"An unexpected error occurred while loading the dataset: {e}")

# Display the first few rows to confirm loading
df.head()
```

```python
print("\n--- Class Distribution ---")
print(df['Class'].value_counts())
print("\n--- Class Distribution (Percentage) ---")
print(df['Class'].value_counts(normalize=True) * 100)
```

```python
print("\n--- Amount Distribution ---")
plt.figure(figsize=(10, 6))
sns.histplot(df['Amount'], bins=50, kde=True, color='skyblue')
plt.title('Distribution of Transaction Amounts')
plt.xlabel('Amount')
plt.ylabel('Frequency')
plt.show()

# Compare Amount distribution for Normal vs. Fraud
plt.figure(figsize=(12, 6))
sns.histplot(df[df['Class'] == 0]['Amount'], bins=50, color='blue', label='Normal (Class 0)', kde=True, alpha=0.6)
sns.histplot(df[df['Class'] == 1]['Amount'], bins=50, color='red', label='Fraud (Class 1)', kde=True, alpha=0.6)
plt.title('Transaction Amount Distribution by Class')
plt.xlabel('Amount')
plt.ylabel('Frequency')
plt.legend()
plt.yscale('log') # Use log scale for y-axis to see fraud transactions better
plt.show()

print("\n--- Time Distribution ---")
plt.figure(figsize=(10, 6))
sns.histplot(df['Time'], bins=50, kde=True, color='lightgreen')
plt.title('Distribution of Transaction Time')
plt.xlabel('Time (seconds since first transaction)')
plt.ylabel('Frequency')
plt.show()

# Time distribution for Normal vs. Fraud
plt.figure(figsize=(12, 6))
sns.histplot(df[df['Class'] == 0]['Time'], bins=50, color='blue', label='Normal (Class 0)', kde=True, alpha=0.6)
sns.histplot(df[df['Class'] == 1]['Time'], bins=50, color='red', label='Fraud (Class 1)', kde=True, alpha=0.6)
plt.title('Transaction Time Distribution by Class')
plt.xlabel('Time (seconds since first transaction)')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

# Create a copy to work on, leaving the original df untouched
df_processed = df.copy()

# Scale 'Amount'
scaler_amount = StandardScaler()
df_processed['Amount_Scaled'] = scaler_amount.fit_transform(df_processed['Amount'].values.reshape(-1, 1))

# Scale 'Time'
scaler_time = StandardScaler()
df_processed['Time_Scaled'] = scaler_time.fit_transform(df_processed['Time'].values.reshape(-1, 1))

# Drop original 'Time' and 'Amount' columns, and the original 'Class' for now
# We'll re-add 'Class' as the target variable later
df_processed = df_processed.drop(['Time', 'Amount'], axis=1)

# Reorder columns to place 'Class' at the end for convenience
# Get all columns except 'Class'
cols = [col for col in df_processed.columns if col != 'Class']
# Add 'Class' to the end
df_processed = df_processed[cols + ['Class']]

print("\n--- Processed DataFrame Head ---")
df_processed.head()

# Save the processed DataFrame to the 'data/processed/' directory
# Exclude the index column when saving
df_processed.to_csv('data/processed/processed_transactions.csv', index=False)
print("\nProcessed data saved to data/processed/processed_transactions.csv")
```
